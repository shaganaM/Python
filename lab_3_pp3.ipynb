```python
def count_vowels(s):
    vowels = "aeiou"
    return sum(1 for ch in s.lower() if ch in vowels)
print(count_vowels("Hello World"))
```

```python
students = {
    "shagana": [78, 85, 90],
    "Bala": [60, 72, 68],
    "Chandra": [88, 91, 84],
    "surya": [45, 50, 40]
}
def calculate_average(marks):
    return sum(marks) / len(marks)
result = {}
for name, marks in students.items():
    avg = round(calculate_average(marks), 2)
    status = (lambda x: "Pass" if x >= 50 else "Fail")(avg)
    result[name] = (avg, status)
for name, (avg, status) in result.items():
    if status == "Pass":
        print(name, avg, status)
topper = max(result, key=lambda x: result[x][0])
print("Topper:", topper)
```

```python
def count_digits(n):
    return 1 if n < 10 else 1 + count_digits(n // 10)
def sum_digits(n):
    return n if n < 10 else n % 10 + sum_digits(n // 10)
def max_digit(n):
    return n if n < 10 else max(n % 10, max_digit(n // 10))
def reverse_number(n, rev=0):
    return rev if n == 0 else reverse_number(n // 10, rev * 10 + n % 10)
num = 4729
print(count_digits(num))
print(sum_digits(num))
print(max_digit(num))
print(reverse_number(num))
```
