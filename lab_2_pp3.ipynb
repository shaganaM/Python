```python
students = ["shagana", "chandra", "surya", "boobi", "thillai"]
print("Initial list:", students)
students.append("Eve")
print("After append:", students)
backup_students = students.copy()
print("Backup list:", backup_students)
boobi_count = students.count("boobi")
print("Count of 'boobi':", boobi_count)
transfer_students = ["saran", "ashwin"]
students.extend(transfer_students)
print("After merging transfer students:", students)
surya_position = students.index("surya")
print("Position of 'surya':", surya_position)
students.insert(1, "vegi")
print("After inserting 'vegi':", students)
removed_last_student = students.pop()
print("Removed last student:", removed_last_student)
print("After pop:", students)
students.remove("chandra")
print("After removing 'chandra':", students)
reversed_students = students[::-1]
print("Reversed list:", reversed_students)
students.sort()
print("Sorted list:", students)
students.clear()
print("After clearing list:", students)

```
